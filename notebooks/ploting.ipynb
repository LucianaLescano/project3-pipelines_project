```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Ploting

## Kaggle data

```python
kaggle_data = pd.read_csv('../output/kaggle_dataset.csv', index_col='Unnamed: 0')
```

```python
kaggle_data
```

```python
one = sns.scatterplot(x='Deaths', y='Year', data=kaggle_data, label='United States')
```

```python
one.figure.savefig("whole_usa.jpg", dpi=1000)
```

## Api data

```python
api_data = pd.read_csv('../output/api_dataset.csv', index_col='Unnamed: 0')
```

```python
api_data
```

```python
a = api_data[api_data.State.isin(['New York'])]
sns.scatterplot(data=a, x='Deaths', y='Year', label='New York', s=80, color='0.695')
```

```python
b = api_data[api_data.State.isin(['California'])]
sns.scatterplot(data=a, x='Deaths', y='Year', label='California', s=80, color='0.695')
```

```python
plt.figure(figsize=(11,7))
two = sns.scatterplot(x='Deaths', y='State', data=api_data, hue='Year', palette="Reds")
```

```python
two.figure.savefig("all_states.jpg", dpi=1000)
```

```python
sns.scatterplot(x='Year', y='Deaths', data=api_data, hue='State', s=80)
```

```python
sns.catplot(x='Deaths', y='State', data=api_data)
```

```python
sns.pairplot(data=api_data)
```

```python
g = sns.FacetGrid(data=api_data, row='State', sharex=False)
three = g.map_dataframe(sns.scatterplot, x='Deaths', y='Year')
```

```python
three.savefig("eachs_state.jpg", dpi=1000)
```
